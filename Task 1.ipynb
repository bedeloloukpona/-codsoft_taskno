```python
import tkinter as tk
from tkinter import messagebox, simpledialog

class ToDoApp:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List")
        
        # Liste des tâches
        self.tasks = []
        
        # Interface utilisateur
        self.task_input = tk.Entry(root, width=50)
        self.task_input.pack(pady=10)
        
        self.add_button = tk.Button(root, text="Ajouter tâche", command=self.add_task)
        self.add_button.pack(pady=5)
        
        self.task_listbox = tk.Listbox(root, width=50, height=15)
        self.task_listbox.pack(pady=10)
        
        self.modify_button = tk.Button(root, text="Modifier tâche", command=self.modify_task)
        self.modify_button.pack(pady=5)
        
        self.delete_button = tk.Button(root, text="Supprimer tâche", command=self.delete_task)
        self.delete_button.pack(pady=5)
        
        self.clear_button = tk.Button(root, text="Effacer toutes les tâches", command=self.clear_tasks)
        self.clear_button.pack(pady=5)
    
    def add_task(self):
        task = self.task_input.get()
        if task:
            self.tasks.append(task)
            self.update_task_list()
            self.task_input.delete(0, tk.END)
        else:
            messagebox.showwarning("Avertissement", "La tâche ne peut pas être vide !")
    
    def delete_task(self):
        selected_task_index = self.task_listbox.curselection()
        if selected_task_index:
            del self.tasks[selected_task_index[0]]
            self.update_task_list()
        else:
            messagebox.showwarning("Avertissement", "Veuillez sélectionner une tâche à supprimer.")
    
    def modify_task(self):
        selected_task_index = self.task_listbox.curselection()
        if selected_task_index:
            # Obtenir la tâche sélectionnée
            current_task = self.tasks[selected_task_index[0]]
            
            # Fenêtre popup pour saisir la nouvelle tâche
            new_task = simpledialog.askstring("Modifier la tâche", "Nouvelle description de la tâche :", initialvalue=current_task)
            if new_task:
                self.tasks[selected_task_index[0]] = new_task
                self.update_task_list()
        else:
            messagebox.showwarning("Avertissement", "Veuillez sélectionner une tâche à modifier.")
    
    def clear_tasks(self):
        self.tasks = []
        self.update_task_list()
    
    def update_task_list(self):
        self.task_listbox.delete(0, tk.END)
        for task in self.tasks:
            self.task_listbox.insert(tk.END, task)

if __name__ == "__main__":
    root = tk.Tk()
    app = ToDoApp(root)
    root.mainloop()
```
