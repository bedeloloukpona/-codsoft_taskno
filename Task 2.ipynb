```python
import tkinter as tk
from tkinter import messagebox

# Fonction pour effectuer les calculs
def calculer():
    try:
        num1 = float(entree1.get())  # Récupérer le premier nombre
        num2 = float(entree2.get())  # Récupérer le deuxième nombre
        operation = choix_op.get()   # Récupérer l'opération choisie

        if operation == '+':
            result = num1 + num2
        elif operation == '-':
            result = num1 - num2
        elif operation == '*':
            result = num1 * num2
        elif operation == '/':
            if num2 != 0:
                result = num1 / num2
            else:
                messagebox.showerror("Erreur", "Division par zéro impossible")
                return
        else:
            messagebox.showerror("Erreur", "Opération invalide")
            return

        label_result.config(text=f"Résultat: {result}")
    except ValueError:
        messagebox.showerror("Erreur", "Veuillez entrer des nombres valides.")

# Création de la fenêtre principale
fenetre = tk.Tk()
fenetre.title("Calculatrice")

# Création des widgets
label1 = tk.Label(fenetre, text="Entrez le premier nombre:")
label1.grid(row=0, column=0)

entree1 = tk.Entry(fenetre)
entree1.grid(row=0, column=1)

label2 = tk.Label(fenetre, text="Entrez le deuxième nombre:")
label2.grid(row=1, column=0)

entree2 = tk.Entry(fenetre)
entree2.grid(row=1, column=1)

label3 = tk.Label(fenetre, text="Choisissez une opération:")
label3.grid(row=2, column=0)

choix_op = tk.StringVar(value='+')
operateurs = ['+', '-', '*', '/']
menu_op = tk.OptionMenu(fenetre, choix_op, *operateurs)
menu_op.grid(row=2, column=1)

# Bouton de calcul
button_calculer = tk.Button(fenetre, text="Calculer", command=calculer)
button_calculer.grid(row=3, column=0, columnspan=2)

# Affichage du résultat
label_result = tk.Label(fenetre, text="Résultat: ")
label_result.grid(row=4, column=0, columnspan=2)

# Lancer l'interface graphique
fenetre.mainloop()
```
