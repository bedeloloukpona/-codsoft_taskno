```python
import tkinter as tk
from tkinter import messagebox
import random
import string

# Fonction pour générer un mot de passe
def generer_mot_de_passe():
    try:
        # Récupérer la longueur du mot de passe et la complexité
        longueur = int(entree_longueur.get())
        complexite = var_complexite.get()

        # Définir les types de caractères à utiliser selon la complexité choisie
        if complexite == "1":
            caracteres = string.ascii_lowercase  # Lettres minuscules uniquement
        elif complexite == "2":
            caracteres = string.ascii_letters  # Lettres minuscules et majuscules
        elif complexite == "3":
            caracteres = string.ascii_letters + string.digits  # Lettres + chiffres
        elif complexite == "4":
            caracteres = string.ascii_letters + string.digits + string.punctuation  # Lettres + chiffres + symboles
        else:
            messagebox.showerror("Erreur", "Option de complexité invalide")
            return

        # Générer un mot de passe aléatoire
        mot_de_passe = ''.join(random.choice(caracteres) for _ in range(longueur))
        
        # Afficher le mot de passe dans le label
        label_resultat.config(text=f"Mot de passe généré : {mot_de_passe}")

    except ValueError:
        messagebox.showerror("Erreur", "Veuillez entrer une longueur valide.")

# Fenêtre principale
fenetre = tk.Tk()
fenetre.title("Générateur de Mot de Passe")

# Widgets pour la longueur du mot de passe
label_longueur = tk.Label(fenetre, text="Longueur du mot de passe :")
label_longueur.pack()

entree_longueur = tk.Entry(fenetre)
entree_longueur.pack()

# Widgets pour la complexité
label_complexite = tk.Label(fenetre, text="Complexité du mot de passe (1-4) :")
label_complexite.pack()

var_complexite = tk.StringVar()
radio1 = tk.Radiobutton(fenetre, text="1 - Minuscule", variable=var_complexite, value="1")
radio2 = tk.Radiobutton(fenetre, text="2 - Minuscule + Majuscule", variable=var_complexite, value="2")
radio3 = tk.Radiobutton(fenetre, text="3 - Lettres + Chiffres", variable=var_complexite, value="3")
radio4 = tk.Radiobutton(fenetre, text="4 - Lettres + Chiffres + Symboles", variable=var_complexite, value="4")

radio1.pack()
radio2.pack()
radio3.pack()
radio4.pack()

# Bouton pour générer le mot de passe
button_generer = tk.Button(fenetre, text="Générer le mot de passe", command=generer_mot_de_passe)
button_generer.pack()

# Afficher le mot de passe généré
label_resultat = tk.Label(fenetre, text="Mot de passe généré :")
label_resultat.pack()

# Lancer l'interface graphique
fenetre.mainloop()
```
