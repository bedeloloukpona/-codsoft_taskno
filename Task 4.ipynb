```python
import tkinter as tk
import random

# Variables globales pour le score
score_utilisateur = 0
score_ordinateur = 0

# Fonction pour déterminer le gagnant
def determiner_gagnant(choix_utilisateur, choix_ordinateur):
    global score_utilisateur, score_ordinateur
    
    if choix_utilisateur == choix_ordinateur:
        resultat = "Égalité!"
    elif (choix_utilisateur == "pierre" and choix_ordinateur == "ciseaux") or \
         (choix_utilisateur == "papier" and choix_ordinateur == "pierre") or \
         (choix_utilisateur == "ciseaux" and choix_ordinateur == "papier"):
        resultat = "Vous avez gagné!"
        score_utilisateur += 1
    else:
        resultat = "L'ordinateur a gagné!"
        score_ordinateur += 1

    # Mettre à jour l'affichage du résultat et des scores
    label_resultat.config(text=f"Votre choix: {choix_utilisateur.capitalize()}\nChoix de l'ordinateur: {choix_ordinateur.capitalize()}\n{resultat}")
    label_scores.config(text=f"Score - Vous: {score_utilisateur} | Ordinateur: {score_ordinateur}")

# Fonction pour démarrer un nouveau tour de jeu
def nouveau_tour(choix_utilisateur):
    # Liste des choix possibles
    choix_possibles = ["pierre", "papier", "ciseaux"]
    # Choix aléatoire de l'ordinateur
    choix_ordinateur = random.choice(choix_possibles)
    
    determiner_gagnant(choix_utilisateur, choix_ordinateur)

# Fonction pour afficher un message de confirmation pour rejouer
def demander_rejouer():
    rejouer = messagebox.askyesno("Rejouer", "Voulez-vous jouer un autre tour ?")
    if rejouer:
        label_resultat.config(text="Faites un choix: Pierre, Papier ou Ciseaux")
        bouton_rejouer.pack_forget()
        bouton_pierre.pack()
        bouton_papier.pack()
        bouton_ciseaux.pack()

# Création de la fenêtre principale
fenetre = tk.Tk()
fenetre.title("Jeu Pierre-Papier-Ciseaux")

# Texte de bienvenue et d'instructions
label_instructions = tk.Label(fenetre, text="Bienvenue dans le jeu Pierre-Papier-Ciseaux!\nChoisissez Pierre, Papier ou Ciseaux pour commencer.", font=("Arial", 14))
label_instructions.pack()

# Boutons pour les choix de l'utilisateur
bouton_pierre = tk.Button(fenetre, text="Pierre", command=lambda: nouveau_tour("pierre"), width=20)
bouton_pierre.pack()

bouton_papier = tk.Button(fenetre, text="Papier", command=lambda: nouveau_tour("papier"), width=20)
bouton_papier.pack()

bouton_ciseaux = tk.Button(fenetre, text="Ciseaux", command=lambda: nouveau_tour("ciseaux"), width=20)
bouton_ciseaux.pack()

# Label pour afficher les résultats et scores
label_resultat = tk.Label(fenetre, text="Faites un choix: Pierre, Papier ou Ciseaux", font=("Arial", 12))
label_resultat.pack()

label_scores = tk.Label(fenetre, text="Score - Vous: 0 | Ordinateur: 0", font=("Arial", 12))
label_scores.pack()

# Bouton "Rejouer" (initialement caché)
bouton_rejouer = tk.Button(fenetre, text="Rejouer", command=demander_rejouer, width=20)

# Lancer l'interface graphique
fenetre.mainloop()
```
